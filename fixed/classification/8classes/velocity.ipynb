```python
import pandas as pd
import numpy as np
import math
```

```python
# read dataset
dataset_path = '/home/lab606a/ML/trajectories/fixed/classification/8classes/test/10balls/top_speed6_20200304_test.csv'
dataset = pd.read_csv(dataset_path, header=None)
dataset = np.array(dataset)
print(dataset.shape)
```

```python
v = np.zeros([250,1])
v.shape
```

```python
vx = 0
vy = 0
vz = 0
for i in range(dataset.shape[0]):
    vx = 0
    vy = 0
    vz = 0
    for j in range(24):
        vx = vx + (dataset[i,j+3] - dataset[i,j])**2
        vy = vy + (dataset[i,j+4] - dataset[i,j+1])**2
        vz = vz + (dataset[i,j+5] - dataset[i,j+2])**2
    v[i,:] = math.sqrt(vx+vy+vz)/250
    #print("i = ", i, v)
print(v)
```

```python
df = pd.DataFrame(data=v)
df
```

```python
csv_path = './velocity.csv'
df.to_csv(csv_path)
```
